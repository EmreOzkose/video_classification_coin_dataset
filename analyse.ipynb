```python
import matplotlib.pyplot as plt
```

```python
exp_path = "exps/exp1"
log_path = "exps/exp1/training.log"
```

```python
train_loss_list, val_loss_list = [], []
train_acc_list, val_acc_list = [], []

for line in open(log_path).read().splitlines():
    if "train Loss" in line:
        _, _, _, _, _, loss, _, acc = line.split(" ")
        train_loss_list.append(float(loss))
        train_acc_list.append(float(acc))

    if "val Loss" in line:
        _, _, _, _, _, loss, _, acc = line.split(" ")
        val_loss_list.append(float(loss))
        val_acc_list.append(float(acc))
```

```python
plt.plot(train_acc_list, label="train")
plt.plot(val_acc_list, label="val")

plt.title("Accuracy Graphs")
plt.legend()
plt.show()
```

```python
plt.plot(train_loss_list, label="train")
plt.plot(val_loss_list, label="val")

plt.title("Loss Graphs")
plt.legend()
plt.show()
```
